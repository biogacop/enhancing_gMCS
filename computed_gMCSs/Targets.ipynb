# Blood

```python
lista_gMCS=[]
for i in range(1,58):
    nombre="./gMCS/gMCS_blood/blood_Task"+str(i)+"_CS.mat"
    with open(nombre, "rb") as f:
        datos = pickle.load(f)
        lista_gMCS.append(datos)
```

```python
len(lista_gMCS)
```

```python
lista_essential=[]
lista_letalities=[]
for i in lista_gMCS:
    lista_essential.append(i[1])
    lista_letalities.append(i[2])
```

```python
listaLS=dict()
for mt in lista_letalities:
    for cs in mt:
        if not cs in listaLS:
            listaLS[cs]=set([lista_letalities.index(mt)])
        else:
            listaLS[cs].add(lista_letalities.index(mt))
            
for j in listaLS:
    print(j,len(listaLS[j]))
    
```

```python
lista_mt=dict()
for mt in lista_letalities:
    for cs in mt:
        if not lista_letalities.index(mt) in lista_mt:
            lista_mt[lista_letalities.index(mt)]=set(cs)
        else:
            lista_mt[lista_letalities.index(mt)].add(cs)

'''for j in lista_mt:
    print("Task"+str(j+1),lista_mt[j])
'''
```

```python
blood=set()

for mt in lista_letalities:
    for cs in mt:
        blood.add(cs)
len(blood)
```

```python


x = range(1, 21)
y = []

for i in x:
    cont=0
    for g in listaLS:
        if len(listaLS[g])==i:
            cont+=1
    y.append(cont)

plt.bar(x, y)
plt.xlabel('Nmero de tareas')
plt.ylabel('Frecuencia de genes')
plt.title('Blood')

#plt.savefig("essentialTasks-Human1.jpg", bbox_inches='tight')
plt.show()
#nmero de tareas metablicas que el modelo mutante 
```
